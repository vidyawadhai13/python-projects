```python
# Largest number from a list
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

largest = max(numbers)
print("Largest number is:", largest)
```

```python
# Print list after removing even numbers
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

result = [num for num in numbers if num % 2 != 0]
print("List after removing even numbers:", result)
```

```python
# Reverse a list
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

numbers.reverse()
print("Reversed list:", numbers)
```

```python
# Square each element of a list
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

squared = [num ** 2 for num in numbers]
print("Squared list:", squared)
```

```python
# Remove empty element from a list
num = ["Hello", 34, 45, "", 40]

result = [item for item in num if item != ""]
print("Result:", result)
```

```python
# Display items divisible by 5
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

result = [num for num in numbers if num % 5 == 0]
print("Numbers divisible by 5:", result)
```

```python
# Remove repetitive items from a list
num = [2, 3, 4, 5, 2, 6, 3, 2]

result = list(set(num))
result.sort()
print("Result:", result)
```

```python
# Append second list to first list
list1 = [23, 24, 25, 26]
list2 = [27, 28, 29, 30]

list1.extend(list2)
print("Result:", list1)
```

```python
# Filter odd and even numbers from a list
numbers = [2, 23, 24, 51, 46, 67]

even = [num for num in numbers if num % 2 == 0]
odd = [num for num in numbers if num % 2 != 0]

print("Even", even)
print("Odd", odd)
```

```python
# Arrange elements in sorted order
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

numbers.sort()
print("Sorted list:", numbers)
```

```python
# Count number of items in a list
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

count = len(numbers)
print("Number of items in the list:", count)
```

```python
# Reverse list without using built-in functions (indexing method)
characters = ['a', 'b', 'c', 'd', 'e']

reversed_list = []
for i in range(len(characters)-1, -1, -1):
    reversed_list.append(characters[i])

print("Reversed list:", reversed_list)
```

```python
# Add item 7000 after 6000 in nested list
list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]

list1[2][2].append(7000)

print("Updated list:", list1)
```

```python
# Replace first occurrence of 20 with 200
list1 = [5, 10, 15, 20, 25, 50, 20]

if 20 in list1:
    index = list1.index(20)
    list1[index] = 200

print("Updated list:", list1)
```

```python
# Find range (difference between largest and smallest element)
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

largest = max(numbers)
smallest = min(numbers)

range_value = largest - smallest

print("Range of the list:", range_value)
```

```python
# Insert a new element at a specific position in the list
# Create list
nums = [10, 20, 30, 40, 50]

# Input from user
pos = int(input("Enter position (0-based index): "))
element = int(input("Enter element to insert: "))

# Insert element
nums.insert(pos, element)

print("Updated list:", nums)
```

```python
# Find the common elements between two lists
# (a)
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7]

common = list(set(list1) & set(list2))

print("Common elements:", common)
```

```python
# (b)
list1 = list(map(int, input("Enter first list elements: ").split()))
list2 = list(map(int, input("Enter second list elements: ").split()))

common = list(set(list1) & set(list2))

print("Common elements:", common)
```

```python
# Find the second smallest element
nums = list(map(int, input("Enter list elements: ").split()))

unique_nums = list(set(nums))
unique_nums.sort()

if len(unique_nums) >= 2:
    print("Second smallest element:", unique_nums[1])
else:
    print("Not enough elements")
```

```python
# Remove elements at even indices
nums = list(map(int, input("Enter list elements: ").split()))

result = [nums[i] for i in range(len(nums)) if i % 2 != 0]

print("List after removing even index elements:", result)
```

```python
# Sum of elements at odd indices
nums = list(map(int, input("Enter list elements: ").split()))

sum_odd_index = sum(nums[i] for i in range(len(nums)) if i % 2 != 0)

print("Sum of elements at odd indices:", sum_odd_index)
```
