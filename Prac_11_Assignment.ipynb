```python
# Lab Assignment 1 – Cosmetic Sales Analysis
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (you can replace with CSV if provided)
data = {
    "month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "facecream": [2500,2630,2140,3400,3600,2760,2980,3700,3540,1990,2340,2900],
    "facewash": [1500,1200,1340,1130,1740,1555,1120,1780,1890,2100,1760,2300],
    "toothpaste": [5200,5100,4550,5870,6000,5250,6100,8300,7300,6200,5800,6100],
    "bathingsoap": [9200,6100,9550,8870,9960,8100,10300,13300,14400,11500,9800,12300],
    "shampoo": [1200,2100,3550,1870,1560,1890,1780,2100,2300,2400,1800,1900],
    "moisturizer": [1500,1200,1340,1130,1740,1555,1120,1780,1890,2100,1760,2300],
    "total_profit": [211000,183300,224400,256400,267800,232900,244400,281400,275400,260300,248900,275500]
}

df = pd.DataFrame(data)

# a) Line Plot of Total Profit
plt.figure()
plt.plot(df["month"], df["total_profit"], marker='o')
plt.title("Total Profit per Month")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid()
plt.show()


# b) Multiline Plot for all products
plt.figure()
for col in ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]:
    plt.plot(df["month"], df[col], marker='o', label=col)

plt.title("Sales Data of All Products")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid()
plt.show()


# c) Bar Chart for Facecream & Facewash
x = range(len(df["month"]))

plt.figure()
plt.bar(x, df["facecream"], width=0.4, label="Facecream")
plt.bar([i + 0.4 for i in x], df["facewash"], width=0.4, label="Facewash")

plt.xticks([i + 0.2 for i in x], df["month"])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Facecream vs Facewash Sales")
plt.legend()
plt.show()


# d) Pie Chart of total yearly sales
total_sales = df.drop(columns=["month","total_profit"]).sum()

plt.figure()
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%')
plt.title("Total Sales Distribution (Yearly)")
plt.show()
```

```python
# Lab Assignment 2 – Recruitment Data Visualization
import matplotlib.pyplot as plt

# Sample recruitment dataset
companies = ["Microsoft","Google","Amazon","IBM","Deloitte","Capgemini","ATOS","Amdocs"]
recruitments = [120, 150, 180, 90, 110, 130, 70, 95]

# a) Bar Chart
plt.figure()
plt.bar(companies, recruitments)
plt.title("Recruitments in Companies")
plt.xlabel("Companies")
plt.ylabel("Number of Recruitments")
plt.xticks(rotation=30)
plt.show()


# b) Pie Chart
plt.figure()
plt.pie(recruitments, labels=companies, autopct='%1.1f%%')
plt.title("Recruitment Share")
plt.show()


# c) Customized Pie Chart
plt.figure()
explode = [0.1 if i == max(recruitments) else 0 for i in recruitments]

plt.pie(recruitments, labels=companies, autopct='%1.1f%%',
        explode=explode, shadow=True, startangle=140)
plt.title("Customized Recruitment Pie Chart")
plt.show()


# d) Doughnut Chart
plt.figure()
plt.pie(recruitments, labels=companies, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Doughnut Chart")
plt.show()


# e) Compare IBM vs Amdocs
ibm = recruitments[companies.index("IBM")]
amdocs = recruitments[companies.index("Amdocs")]

plt.figure()
plt.bar(["IBM","Amdocs"], [ibm, amdocs])
plt.title("IBM vs Amdocs Recruitment Comparison")
plt.ylabel("Number of Recruitments")
plt.show()
```
