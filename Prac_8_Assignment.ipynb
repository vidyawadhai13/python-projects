```python
# Lab Assignment 1
# Read a text file and write its contents to a new file in UPPERCASE

# Create the source file and ask user for text
text = input("Type the text you want to store in the file:\n")

# write this text into a file named input.txt
with open("input.txt", "w", encoding="utf-8") as f:
    f.write(text)

# Read from input.txt, write uppercase to output.txt
src = "input.txt"
dst = "output.txt"

with open(src, "r", encoding="utf-8") as fin, \
     open(dst, "w", encoding="utf-8") as fout:
    for line in fin:
        fout.write(line.upper())

# Show contents of both files
print("\nContent of input.txt: ")
with open(src, "r", encoding="utf-8") as f:
    print(f.read())

print("\nContent of output.txt (uppercase): ")
with open(dst, "r", encoding="utf-8") as f:
    print(f.read())
```

```python
# Lab Assignment 2

# Copy Python script to another file without comments

# Create a sample Python source file with comments
print("Type your Python code (press Enter twice when done):")
lines = []
while True:
    line = input()
    if line == "" and len(lines) > 0 and lines[-1] == "":
        break
    lines.append(line)

# Save source code to source.py
with open("source.py", "w", encoding="utf-8") as f:
    f.write("\n".join(lines) + "\n")

# Copy source.py to no_comments.py (excluding comments)
with open("source.py", "r", encoding="utf-8") as fin, \
     open("no_comments.py", "w", encoding="utf-8") as fout:

    for line in fin:
        stripped = line.lstrip()
        # Skip lines that start with # (full line comments) or empty lines
        if stripped.startswith("#") or stripped.strip() == "":
            continue
        fout.write(line)

# Show contents of both files
print("SOURCE FILE - WITH COMMENTS:")
with open("source.py", "r", encoding="utf-8") as f:
    print(f.read())

print("DESTINATION FILE - WITHOUT COMMENTS:")
with open("no_comments.py", "r", encoding="utf-8") as f:
    print(f.read())
```
