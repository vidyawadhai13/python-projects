```python
import heapq

campus_graph = {
    "Main Gate": {"Library": 40, "Parking": 30},
    "Library": {"Main Gate": 40, "Admin Office": 50, "Canteen": 30},
    "Canteen": {"Library": 30, "Hostel": 40},
    "Admin Office": {"Library": 50, "Computer Lab": 20},
    "Computer Lab": {"Admin Office": 20, "Auditorium": 60},
    "Auditorium": {"Computer Lab": 60, "Parking": 50},
    "Hostel": {"Canteen": 40},
    "Parking": {"Main Gate": 30, "Auditorium": 50}
}

def dijkstra(graph, start, end):
    queue = [(0, start, [])]
    visited = set()
    while queue:
        distance, node, path = heapq.heappop(queue)
        if node in visited: continue
        visited.add(node)
        path = path + [node]
        if node == end: return distance, path
        for neighbour, weight in graph[node].items():
            if neighbour not in visited:
                heapq.heappush(queue, (distance + weight, neighbour, path))
    return float("inf"), []

print("--- Campus Navigation System (Distances in Meters) ---")
print("Available Locations:", list(campus_graph.keys()))

while True:
    start = input("Enter starting location (or 'quit'): ")
    if start.lower() == 'quit': break
    end = input("Enter destination location: ")
    distance, path = dijkstra(campus_graph, start, end)
    if distance != float("inf"):
        print("Shortest Path:", " -> ".join(path))
        print("Total Distance:", f"{distance} meters")
    else:
        print("No path found!")
```
