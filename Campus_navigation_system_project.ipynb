```python
import heapq

# Graph representing campus locations and distances
campus_graph = {
    "Main Gate": {"Library": 4, "Parking": 3},
    "Library": {"Main Gate": 4, "Admin Office": 5, "Canteen": 3},
    "Canteen": {"Library": 3, "Hostel": 4},
    "Admin Office": {"Library": 5, "Computer Lab": 2},
    "Computer Lab": {"Admin Office": 2, "Auditorium": 6},
    "Auditorium": {"Computer Lab": 6, "Parking": 5},
    "Hostel": {"Canteen": 4},
    "Parking": {"Main Gate": 3, "Auditorium": 5}
}

def dijkstra(graph, start, end):
    queue = [(0, start, [])]
    visited = set()

    while queue:
        distance, node, path = heapq.heappop(queue)

        if node in visited:
            continue

        visited.add(node)
        path = path + [node]

        if node == end:
            return distance, path

        for neighbour, weight in graph[node].items():
            if neighbour not in visited:
                heapq.heappush(queue, (distance + weight, neighbour, path))

    return float("inf"), []


print("\n--- Campus Navigation System ---\n")

print("Available Locations:")
for location in campus_graph:
    print("-", location)

start = input("\nEnter starting location: ")
end = input("Enter destination location: ")

distance, path = dijkstra(campus_graph, start, end)

print("\nShortest Path:", " -> ".join(path))
print("Total Distance:", distance)
```
