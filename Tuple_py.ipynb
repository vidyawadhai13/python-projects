```python
# Create a tuple with three elements
t = (10, "hello", 3.14)
print(t)
```

```python
# Access and print the third element
t = (1, 2, 3, 4, 5)
print("Third element:", t[2])
```

```python
# Concatenate two tuples
t1 = (1, 2, 3)
t2 = (4, 5, 6)

result = t1 + t2
print("Concatenated tuple:", result)
```

```python
# Unpack the tuple into variables
t = ('a', 'b', 'c')
x, y, z = t

print(x, y, z)
```

```python
# Check if element 7 exists
t = (1, 3, 5, 7, 9)

if 7 in t:
    print("7 exists in tuple")
else:
    print("7 does not exist")
```

```python
# Print slice from index 2 to 4
t = (0, 1, 2, 3, 4, 5)
print("Sliced tuple:", t[2:5])
```

```python
# Find the length of the tuple
t = (10, 20, 30, 40, 50)
print("Length:", len(t))
```

```python
# Repeat tuple three times
t = (1, 2, 3)
print("Repeated tuple:", t * 3)
```

```python
# Convert list to tuple
lst = [1, 2, 3]
t = tuple(lst)

print("Tuple:", t)
```

```python
# Find minimum and maximum
t = (5, 12, 3, 8, 15)

print("Minimum:", min(t))
print("Maximum:", max(t))
```

```python
# Count occurrences of element 2
t = (1, 2, 3, 2, 4, 2)
print("Count of 2:", t.count(2))
```

```python
# Check if all elements are True
t = (True, True, False)

if all(t):
    print("All elements are True")
else:
    print("Not all elements are True")
```

```python
# Find index of first occurrence of 3
t = (1, 2, 3, 4, 3, 5)

print("Index of first 3:", t.index(3))
```

```python
# Sort tuple and store in new tuple
t = (5, 2, 8, 1, 3)

sorted_t = tuple(sorted(t))
print("Sorted tuple:", sorted_t)
```

```python
# Tuple of squares from 1 to 5
t = tuple(i*i for i in range(1, 6))
print("Squares tuple:", t)
```

```python
# Python program to create a tuple (user input)
nums = tuple(map(int, input("Enter elements: ").split()))
print("Tuple:", nums)
```

```python
# Tuple with different data types
t = (10, "Python", 3.14, True)
print("Tuple with different data types:", t)
```

```python
# Tuple of numbers and print one item
t = (10, 20, 30, 40)

index = int(input("Enter index to print: "))
print("Element:", t[index])
```

```python
# Unpack tuple into several variables
t = (100, 200, 300)

a, b, c = t
print("a =", a)
print("b =", b)
print("c =", c)
```

```python
# Add an item to a tuple
t = tuple(map(int, input("Enter tuple elements: ").split()))
item = int(input("Enter item to add: "))

t = t + (item,)
print("Updated tuple:", t)
```

```python
# Convert a tuple to a string
t = tuple(input("Enter characters separated by space: ").split())

s = ''.join(t)
print("String:", s)
```

```python
# Get the 4th element from the last
t = tuple(map(int, input("Enter tuple elements: ").split()))

if len(t) >= 4:
    print("4th element from last:", t[-4])
else:
    print("Tuple has less than 4 elements")
```

```python
# Clone (copy) a tuple
t1 = tuple(map(int, input("Enter tuple elements: ").split()))
t2 = t1[:]

print("Original:", t1)
print("Cloned:", t2)
```

```python
# Find repeated items in a tuple
t = tuple(map(int, input("Enter tuple elements: ").split()))

repeated = set([x for x in t if t.count(x) > 1])
print("Repeated items:", repeated)
```

```python
# Check whether an element exists
t = tuple(map(int, input("Enter tuple elements: ").split()))
item = int(input("Enter element to search: "))

if item in t:
    print("Element exists")
else:
    print("Element does not exist")
```

```python
# Convert a list to a tuple
lst = list(map(int, input("Enter list elements: ").split()))
t = tuple(lst)

print("Tuple:", t)
```

```python
# Remove an item from a tuple
t = tuple(map(int, input("Enter tuple elements: ").split()))
item = int(input("Enter element to remove: "))

temp = list(t)
if item in temp:
    temp.remove(item)

t = tuple(temp)
print("Updated tuple:", t)
```

```python
# Slice a tuple
t = tuple(map(int, input("Enter tuple elements: ").split()))

start = int(input("Enter start index: "))
end = int(input("Enter end index: "))

print("Sliced tuple:", t[start:end])
```

```python
# Find the index of an item
t = tuple(map(int, input("Enter tuple elements: ").split()))
item = int(input("Enter element to find index: "))

if item in t:
    print("Index:", t.index(item))
else:
    print("Element not found")
```

```python
# Find the length of a tuple
t = tuple(map(int, input("Enter tuple elements: ").split()))
print("Length:", len(t))
```

```python
# Convert a tuple to a dictionary
n = int(input("Enter number of key-value pairs: "))
lst = []

for _ in range(n):
    k = input("Enter key: ")
    v = input("Enter value: ")
    lst.append((k, v))

t = tuple(lst)
d = dict(t)

print("Dictionary:", d)
```

```python
# Unzip a list of tuples
n = int(input("Enter number of tuples: "))
lst = []

for _ in range(n):
    a, b = map(int, input("Enter two values: ").split())
    lst.append((a, b))

x, y = zip(*lst)

print("List 1:", list(x))
print("List 2:", list(y))
```

```python
# Reverse a tuple
t = tuple(map(int, input("Enter tuple elements: ").split()))

rev = t[::-1]
print("Reversed tuple:", rev)
```

```python
# Convert list of tuples into a dictionary
n = int(input("Enter number of pairs: "))
lst = []

for _ in range(n):
    k = input("Enter key: ")
    v = input("Enter value: ")
    lst.append((k, v))

d = dict(lst)
print("Dictionary:", d)
```

```python
# Remove empty tuples from a list
lst = [(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d',)]

result = [x for x in lst if x]
print("After removing empty tuples:", result)
```
