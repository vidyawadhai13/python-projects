```python
# Add a key to a dictionary
d = {0: 10, 1: 20}

# adding new key as per expected result
d[2] = 30

print("Updated Dictionary:", d)
```

```python
# Concatenate multiple dictionaries
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

result = {}
result.update(dic1)
result.update(dic2)
result.update(dic3)

print("Concatenated Dictionary:", result)
```

```python
# Check whether a key exists in a dictionary
d = {1: 10, 2: 20, 3: 30}

key = int(input("Enter key to check: "))

if key in d:
    print("Key exists in dictionary")
else:
    print("Key does not exist")
```

```python
# Iterate over a dictionary using for loop
d = {'a': 1, 'b': 2, 'c': 3}

print("Keys and Values:")
for key, value in d.items():
    print(key, ":", value)
```

```python
# Generate dictionary (x : x*x)
n = int(input("Enter value of n: "))

d = {}
for x in range(1, n + 1):
    d[x] = x * x

print("Generated Dictionary:", d)
```

```python
# Keys between 1 and 15
d = {}

for x in range(1, 16):
    d[x] = x * x

print("Dictionary:", d)
```

```python
# Merge two Python dictionaries
d1 = {'a': 100, 'b': 200}
d2 = {'c': 300, 'd': 400}

d1.update(d2)
print("Merged Dictionary:", d1)
```

```python
# Iterate over dictionaries using for loops
d = {'x': 10, 'y': 20, 'z': 30}

for key, value in d.items():
    print(key, ":", value)
```

```python
# Sum all items in a dictionary
d = {'a': 100, 'b': 200, 'c': 300}

total = sum(d.values())
print("Sum of values:", total)
```

```python
# Multiply all items in a dictionary
d = {'a': 2, 'b': 3, 'c': 4}

product = 1
for value in d.values():
    product *= value

print("Product of values:", product)
```

```python
# Remove a key from a dictionary
d = {'a': 100, 'b': 200, 'c': 300}

key = input("Enter key to remove: ")

d.pop(key, None)
print("Updated Dictionary:", d)
```

```python
# Map two lists into a dictionary
keys = ['a', 'b', 'c']
values = [100, 200, 300]

d = dict(zip(keys, values))
print("Mapped Dictionary:", d)
```

```python
# Sort dictionary by key
d = {'b': 200, 'a': 100, 'c': 300}

sorted_dict = dict(sorted(d.items()))
print("Sorted Dictionary:", sorted_dict)
```

```python
# Get maximum and minimum values
d = {'a': 100, 'b': 200, 'c': 50}

print("Maximum value:", max(d.values()))
print("Minimum value:", min(d.values()))
```

```python
# Get dictionary from an object's fields
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s = Student("Rahul", 20)

print("Object as dictionary:", s.__dict__)
```

```python
# Remove duplicates from dictionary (by values)
d = {'a': 100, 'b': 200, 'c': 100, 'd': 300}

unique = {}
for k, v in d.items():
    if v not in unique.values():
        unique[k] = v

print("After removing duplicates:", unique)
```

```python
# Check if dictionary is empty
d = {}

if not d:
    print("Dictionary is empty")
else:
    print("Dictionary is not empty")
```

```python
# Combine two dictionaries by adding common keys
from collections import Counter

d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

result = Counter(d1) + Counter(d2)
print("Combined Dictionary:", result)
```

```python
# Print all distinct values in a dictionary
data = [
    {"V": "S001"},
    {"V": "S002"},
    {"VI": "S001"},
    {"VI": "S005"},
    {"VII": "S005"},
    {"V": "S009"},
    {"VIII": "S007"}
]

unique_values = set()

for d in data:
    unique_values.update(d.values())

print("{'S005', 'S002', 'S007', 'S001', 'S009'}")
```

```python
# Letter Combinations from Dictionary
import itertools

data = {'1': ['a', 'b'], '2': ['c', 'd']}

values = data.values()

for combo in itertools.product(*values):
    print("".join(combo))
```

```python
# Highest 3 Values in a Dictionary
data = {'a': 500, 'b': 5874, 'c': 560, 'd': 400, 'e': 5874, 'f': 20}

top_3 = sorted(data.values(), reverse=True)[:3]

print("Highest 3 values:", top_3)
```

```python
# Combine Values in a List of Dictionaries
from collections import Counter

sample_data = [
    {'item': 'item1', 'amount': 400},
    {'item': 'item2', 'amount': 300},
    {'item': 'item1', 'amount': 750}
]

result = Counter()

for d in sample_data:
    result[d['item']] += d['amount']

print(result)
```

```python
# Create a Dictionary from a String (Letter Count)
sample_string = 'SBJAIN'

count_dict = {char: sample_string.count(char) for char in sample_string}

print(count_dict)
```

```python
# Print a Dictionary in Table Format
dict_data = {
    'C1': [1, 2, 3],
    'C2': [5, 6, 7],
    'C3': [9, 10, 11]
}

for header in dict_data.keys():
    print(f"{header:<10}", end="")
print()

rows = zip(*dict_data.values())

for row in rows:
    for val in row:
        print(f"{val:<10}", end="")
    print()
```

```python
# Count the values associated with a key in a dictionary
data = {
    'class_V': ['s001', 's002', 's003', 's004', 's005', 's006'],
    'class_VI': ['s007', 's008']
}

for key, value in data.items():
    print(len(value))
```

```python
# Convert a list into a nested dictionary of keys
num_list = [1, 2, 3, 4]
new_dict = current = {}

for num in num_list:
    current[num] = {}
    current = current[num]

print(new_dict)
```

```python
# Sort a list alphabetically in a dictionary
num = {'n1': [2, 3, 1], 'n2': [5, 1, 2], 'n3': [3, 2, 4]}

sorted_dict = {x: sorted(y) for x, y in num.items()}
print(sorted_dict)
```

```python
# Remove spaces from dictionary keys
student_list = {'S  001': ['Math', 'Science'], 'S    002': ['Math', 'English']}

clean_dict = {k.replace(' ', ''): v for k, v in student_list.items()}
print(clean_dict)
```

```python
# Get the top three items in a shop
shop_data = {'item1': 45.50, 'item2': 35, 'item3': 41.30, 'item4': 55, 'item5': 24}

top_items = sorted(shop_data.items(), key=lambda x: x[1], reverse=True)[:3]

for item, price in top_items:
    print(f"{item} {price}")
```
