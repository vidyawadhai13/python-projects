```python
import pandas as pd

# Create sample books data
data = {
    'Title': ['Book A', 'Book B', 'Book C', 'Book D', 'Book E'],
    'Author': ['R.K. Narayan', 'Chetan Bhagat', 'R.K. Narayan', 'Arundhati Roy', 'Chetan Bhagat'],
    'Edition': [1, 2, 1, 3, 2],
    'Year': [2000, 2010, 2005, 2015, 2020],
    'Publisher': ['ABC', 'XYZ', 'ABC', 'PQR', 'XYZ'],
    'Price': [300, 500, 250, 600, 400]
}

df = pd.DataFrame(data)

# a) Complete report
print("----- Complete Book Report -----")
print(df)

# b) Books by given author
author = input("\nEnter Author Name: ")
print("\nBooks by Author:")
print(df[df['Author'] == author])

# c) Books by given publisher
pub = input("\nEnter Publisher Name: ")
print("\nBooks by Publisher:")
print(df[df['Publisher'] == pub])

# d) Cheapest and costliest book
print("\nCheapest Book:")
print(df[df['Price'] == df['Price'].min()])

print("\nCostliest Book:")
print(df[df['Price'] == df['Price'].max()])

# e) Sort by year
print("\nBooks Sorted by Year:")
print(df.sort_values(by='Year'))
```

```python
import pandas as pd

# Create sample states data
data = {
    'State': ['Maharashtra', 'Gujarat', 'Rajasthan', 'Karnataka', 'Tamil Nadu'],
    'Area': [307713, 196244, 342239, 191791, 130058],   # in sq km
    'Population': [124000000, 70000000, 81000000, 68000000, 78000000]
}

df = pd.DataFrame(data)

# a) Complete information
print("----- State Information -----")
print(df)

# b) State with largest area
max_area = df[df['Area'] == df['Area'].max()]
print("\nState with Largest Area:")
print(max_area['State'].values[0])

# c) State with largest population
max_pop = df[df['Population'] == df['Population'].max()]
print("\nState with Largest Population:")
print(max_pop['State'].values[0])

# d) Calculate population density
df['Density'] = df['Population'] / df['Area']
print("\nPopulation Density:")
print(df[['State', 'Density']])

# e) State with highest population density
max_density = df[df['Density'] == df['Density'].max()]
print("\nState with Highest Population Density:")
print(max_density['State'].values[0])
```
