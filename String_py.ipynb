```python
# Reverse a string (5 different ways)
s = input("Enter a string: ")

# 1. Slicing
print("Slicing:", s[::-1])

# 2. reversed()
print("reversed():", "".join(reversed(s)))

# 3. Loop
rev = ""
for char in s:
    rev = char + rev
print("Loop:", rev)

# 4. List reverse
l = list(s)
l.reverse()
print("List reverse:", "".join(l))

# 5. Recursion
def reverse_string(text):
    if len(text) == 0:
        return text
    return reverse_string(text[1:]) + text[0]

print("Recursion:", reverse_string(s))
```

```python
# Reverse string using stack and reversed method
s = input("Enter a string: ")

# Stack method
stack = list(s)
rev = ""

while stack:
    rev += stack.pop()

print("Reversed (Stack):", rev)

# reversed()
print("Reversed (reversed method):", "".join(reversed(s)))
```

```python
# Count uppercase and lowercase letters
s = input("Enter a string: ")

upper = lower = 0

for char in s:
    if char.isupper():
        upper += 1
    elif char.islower():
        lower += 1

print("Uppercase letters:", upper)
print("Lowercase letters:", lower)
```

```python
# Count letters and digits
s = input("Enter a string: ")

letters = digits = 0

for char in s:
    if char.isalpha():
        letters += 1
    elif char.isdigit():
        digits += 1

print("Letters:", letters)
print("Digits:", digits)
```

```python
# Find the least frequent character in the string
s = input("Enter a string: ")

freq = {}
for char in s:
    freq[char] = freq.get(char, 0) + 1

least = min(freq, key=freq.get)
print("Least frequent character:", least)
print("Frequency:", freq[least])
```

```python
# Find maximum frequency character
s = input("Enter a string: ")

freq = {}
for char in s:
    freq[char] = freq.get(char, 0) + 1

max_char = max(freq, key=freq.get)
print("Maximum frequency character:", max_char)
print("Frequency:", freq[max_char])
```

```python
# Check whether string is binary or not
s = input("Enter a string: ")

if all(char in '01' for char in s):
    print("String is Binary.")
else:
    print("String is Not Binary.")
```

```python
# Find uncommon words from two strings
s1 = input("Enter first sentence: ")
s2 = input("Enter second sentence: ")

set1 = set(s1.split())
set2 = set(s2.split())

uncommon = set1.symmetric_difference(set2)

print("Uncommon words:", uncommon)
```
