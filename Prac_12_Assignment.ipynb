```python
# (1)
import pandas as pd

# Creating the DataFrame
data = {
    'carat': [0.23, 0.21, 0.23, 0.29, 0.31],
    'cut': ['Ideal', 'Premium', 'Good', 'Premium', 'Good'],
    'color': ['E', 'E', 'E', 'I', 'J'],
    'clarity': ['SI2', 'SI1', 'VS1', 'VS2', 'SI2'],
    'depth': [61.5, 59.8, 56.9, 62.4, 63.3],
    'table': [55.0, 61.0, 65.0, 58.0, 58.0],
    'price': [326, 326, 327, 334, 335],
    'x': [3.95, 3.89, 4.05, 4.20, 4.34],
    'y': [3.98, 3.84, 4.07, 4.23, 4.35],
    'z': [2.43, 2.31, 2.31, 2.63, 2.75]
}

df = pd.DataFrame(data)

# i) Mean price for each cut
print("Mean price for each cut:")
print(df.groupby('cut')['price'].mean())

# ii) Count, minimum and maximum price for each cut
print("\nCount, Min and Max price for each cut:")
print(df.groupby('cut')['price'].agg(['count', 'min', 'max']))

# iii) Average of x, y, z
print("\nAverage values:")
print("Average x:", df['x'].mean())
print("Average y:", df['y'].mean())
print("Average z:", df['z'].mean())
```

```python
# (2)
# Import pandas
import pandas as pd

# Step 1: Create sample employee data (no file needed)
data = {
    'Employee ID': [101, 102, 103, 104, 105],
    'Employee Name': ['Amit', 'Neha', 'Raj', 'Priya', 'Karan'],
    'Department': ['Automotive', 'IT', 'Automotive', 'HR', 'Automotive'],
    'Designation': ['Developer', 'Manager', 'Developer', 'HR', 'Developer']
}

df = pd.DataFrame(data)

# Step 2: Display full data
print("----- Full Employee Data -----")
print(df)

# Step 3: a) Employees in Automotive domain
print("\n----- Employees in Automotive Domain -----")
auto_emp = df[df['Department'] == 'Automotive']
print(auto_emp)

# Step 4: b) Details of employee by ID
emp_id = int(input("\nEnter Employee ID: "))
emp = df[df['Employee ID'] == emp_id]

if not emp.empty:
    print("\n----- Employee Details -----")
    print(emp)
else:
    print("Employee not found")

# Step 5: d) List of Developers
print("\n----- List of Developers -----")
developers = df[df['Designation'] == 'Developer']
print(developers)
```
